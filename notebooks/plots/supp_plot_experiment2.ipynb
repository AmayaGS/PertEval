```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data from the provided files
file_1 = np.load('norman_thresholds_sensitivities_scbert.npy')
file_2 = np.load('norman_thresholds_sensitivities_scgpt.npy')
```

```python
# Extract thresholds and sensitivities
thresholds_1, sensitivities_1 = file_1[0] * 100, file_1[1]
thresholds_2, sensitivities_2 = file_2[0] * 100, file_2[1]

# Set the seaborn style
sns.set(style="white")

# Create the line plot
plt.figure()

# Plot the first dataset with seaborn
sns.lineplot(x=thresholds_1, y=sensitivities_1, marker='', label='scBERT', color='#0072B2')  # Blue

# Plot the second dataset with seaborn
sns.lineplot(x=thresholds_2, y=sensitivities_2, marker='', label='scGPT', color='#009E73')  # Green

# Customize the plot style similar to the provided examples
plt.tick_params(axis='both', which='major', labelsize=10)
plt.xlabel(r'$\tau$', fontsize=9)
plt.ylabel(r'$p(\tau)$', fontsize=9)
plt.legend()
plt.grid(False)

# Save the plot as a 300 dpi PDF
output_file = 'paper_figs/supp_plot_experiment2.pdf'
plt.savefig(output_file, format='pdf', dpi=300)
plt.show()
```
